## Loc ve iLoc

```python
import numpy as np
import pandas as pd
```

```python
dizi = np.random.randint(1,30, size=(10,3))
dizi
```

```python
df = pd.DataFrame(dizi, columns = ["veri1","veri2","veri3"])
df
```

```python
#loc ; tanımlandığı şekli ile seçim yapmak için kullanılır
df.loc[0:3]
```

```python
#iloc ; indeksleme mantığı ile seçim yapar
df.iloc[0:3]
```

loc girilen satırıda alır, iloc o satıra kadar olanları alır

##Pivot Tablolar


```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
titanic.pivot_table("survived",index = "sex", columns= "class")
```

```python
titanic.age.head()
```

```python
age = pd.cut(titanic["age"], [0, 18, 90])
age.head(10)
```

```python
titanic.pivot_table("survived", ["sex", age], "class")
```
